```python
import pandas as pd
import requests
```

```python


req = (
    "https://www.ncei.noaa.gov/access/services/data/v1"
    "?dataset=daily-summaries"
    "&stations=USR0000CBAR"
    "&startDate=2010-01-01"
    "&endDate=2022-12-31"
    "&dataTypes=TMAX,TMIN,PRCP,AWND"
    "&format=json"
)


r = requests.get(req)

data = r.json()
print(data)
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
df[["TMAX", "TMIN"]] = df[["TMAX", "TMIN"]].astype(float)
```

```python
df.head()
```

```python
df["TMAX_F"] = df["TMAX"].mul(0.1).mul(9/5).add(32).round()
df["TMIN_F"] = df["TMIN"].mul(0.1).mul(9/5).add(32).round()
```

```python
df.head()
```

```python
df.to_csv("./data/weather_daily_10-22.csv", index=False)
```
